# REQ#01 - Funções para tarefas especificas

```python
def mais_likes():
	cursor.execute("""SELECT filmes.filme_nome, likes.numero_likes
FROM filmes
INNER JOIN likes ON likes.likes_id = filmes.filme_id
WHERE likes.numero_likes ORDER BY likes.numero_likes DESC
LIMIT 5;""")
	resultado = cursor.fetchall()
	for x in resultado:
		print(x)
        
def menos_likes():
    cursor.execute("""
SELECT filmes.filme_nome, likes.numero_likes
FROM filmes
INNER JOIN likes ON likes.likes_id = filmes.filme_id
WHERE likes.numero_likes ORDER BY likes.numero_likes ASC
LIMIT 5;""")
    resultado = cursor.fetchall()
    for x in resultado:
        print(x)
        
def mais_velhos():
    cursor.execute("""
SELECT filmes.filme_nome, filmes.ano
FROM filmes
WHERE filmes.ano ORDER BY filmes.ano ASC
LIMIT 5;""")
    resultado = cursor.fetchall()
    for x in resultado:
        print(x)  

def comedia():
    cursor.execute("""
SELECT filmes.filme_nome, filmes.nota
FROM filmes
INNER JOIN generos ON generos.genero_id = filmes.filme_id
WHERE generos.genero_nome = "Comedy"  ORDER BY filmes.nota DESC
LIMIT 5;""")
    resultado = cursor.fetchall()
    for x in resultado:
        print(x)  

def FFC():
    cursor.execute("""
SELECT filmes.filme_nome
FROM filmes
INNER JOIN diretores ON diretores.diretor_id = filmes.filme_id
WHERE diretores.diretor_nome = "Francis Ford Coppola"
LIMIT 5;""")
    resultado = cursor.fetchall()
    for x in resultado:
        print(x)
        
def filmes_2000():
    cursor.execute("""
SELECT filmes.filme_nome
FROM filmes
WHERE filmes.ano > 2000 ORDER BY filmes.nota DESC
LIMIT 5;""")
    resultado = cursor.fetchall()
    for x in resultado:
        print(x)
        
def melhores():
    cursor.execute("""
SELECT filmes.filme_nome, filmes.nota
FROM filmes
WHERE filmes.nota ORDER BY filmes.nota DESC
LIMIT 5;""")
    resultado = cursor.fetchall()
    for x in resultado:
        print(x)
        
def n_ingles():
    cursor.execute("""
SELECT filmes.filme_nome, linguas.linguagem
FROM filmes
INNER JOIN linguas ON linguas.lingua_id = filmes.filme_id
WHERE linguas.linguagem != "English" ORDER BY filmes.nota DESC
LIMIT 5;""")
    resultado = cursor.fetchall()
    for x in resultado:
        print(x)
        
def anos90():
    cursor.execute("""
SELECT filmes.filme_nome
FROM filmes
WHERE filmes.ano BETWEEN 1990 AND 2000 ORDER BY filmes.nota DESC
LIMIT 5;""")
    resultado = cursor.fetchall()
    for x in resultado:
        print(x)
        
def filmes_tempo():
    cursor.execute("""
SELECT filmes.filme_nome, filmes.duracao
FROM filmes
WHERE filmes.duracao ORDER BY filmes.duracao DESC
LIMIT 5;""")
    resultado = cursor.fetchall()
    for x in resultado:
        print(x)
```

```python
Docs = [doc1,doc2,doc3,doc4,doc5,doc6,doc7]
```

```python
def vocabulario(Docs):
    return list(set([palavra for sentença in Docs for palavra in sentença[0].split()]))
```

```python
def dicionario_Documentos(Docs):
    pos, neg = [], []
    posicao = 1
    for d in Docs:
        if(d[posicao] == Neg):
            neg.append(d)
        else:
            pos.append(d)
    dic = {}
    dic[Neg], dic[Pos] = neg, pos
    return dic
```

```python
Dicdocs = dicionario_Documentos(Docs)
Dicdocs
```

```python
def contar(palavra, Docs):
    posicao = 0
    return sum([1 for d in Docs for p in d[posicao].split() if p == palavra])
```

```python
contar("falhou", Dicdocs[Neg] )
```

```python
def contar_palavras(Docs):
    posicao = 0
    return sum([1 for d in Docs for p in d[posicao].split()])
```

```python
contar_palavras(Dicdocs[Pos])
```

```python
def priori(Docs, classe):
    posicao = 1 
    return sum([1 for d in Docs if  d[posicao] == classe])/ len(Docs)
```

```python
priori(Docs, Neg)
```

```python
priori(Docs, Pos)
```

```python
def posteriori(palavra, classe, Docsclasse, nVoca):
    return (contar(palavra, Docsclasse) + 1) / (contar_palavras(Docsclasse) + nVoca)
```

```python
classe = Pos
posteriori("melhor", classe, Dicdocs[classe], len(Voca))
```

```python
def prever(sentença, Docs):
    Neg, Pos = 0, 1
    mult_Neg, mult_Pos = 1, 1
    log_priori_neg = log(priori(Docs, Neg))
    log_priori_pos = log(priori(Docs, Pos))
    
    Voca = vocabulario(Docs)
    Dicdocs = dicionario_Documentos(Docs)
    
    for palavra in sentença.split():
        if (palavra not in Voca):
            continue
        classe = Pos 
        mult_Pos *= posteriori(palavra, classe, Dicdocs[classe], len(Voca))
        classe = Neg
        mult_Neg *= posteriori(palavra, classe, Dicdocs[classe], len(Voca))
    log_verosimilhança_Pos = log_priori_pos + log(mult_Pos)
    log_verosimilhança_Neg = log_priori_neg + log(mult_Neg)
    
    return Pos if log_verosimilhança_Pos > log_verosimilhança_Neg else Neg  
```

```python
prever("Filme muito bom adorei", Docs)
```
